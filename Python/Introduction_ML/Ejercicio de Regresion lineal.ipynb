```python
#mejora de set de datos, eliminar filas con vacios
datos_na = datos.dropna() #no modifica el conjunto original
datos_na.info()
```

```python
#Convertir caracteristica categorica a numerica de "ocean proximity"
datos_na["ocean_proximity"]
```

```python
datos_na["ocean_proximity"].value_counts()
```

```python
#dumies / oone-hot encoding. Columna donde asigna un 1 si ese es la categoría. 

dummies =pd.get_dummies(datos_na["ocean_proximity"], dtype=int)
```

```python
datos_na = datos_na.join(dummies)
```

```python
datos_na.head()
```

```python
datos_na = datos_na.drop(["ocean_proximity"], axis=1)
```

```python
datos_na.head()
```

```python
#correlaciones datos_na.corr, pero se grafica para facilitar la interpretación
sb.set(rc={'figure.figsize':(15,8)})
sb.heatmap(datos_na.corr(), annot=True, cmap = "YlGnBu")
#la función .corr utiliza el metodo de correlación de pearson por defecto, pero se puede seleccionar en sus parámetros el metodo de kendall o spearman.
```

```python
#para solo concentrarnos en la variable de valor medio de la casa:
datos_na.corr()["median_house_value"].sort_values(ascending=False)
```
